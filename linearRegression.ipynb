```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
dataset=pd.read_csv('/content/house-prices-advanced-regression-techniques-3.csv')
```

```python
dataset.head()
```

```python
dataset.tail()
```

```python
dataset.isnull().sum()
```

```python
numeric_data = dataset.select_dtypes(include=[np.number]).values.flatten()
print(np.std(numeric_data))
```

```python
dataset.describe()
```

```python
dataset['LotFrontage'] = dataset['LotFrontage'].fillna(dataset['LotFrontage'].median())
```

```python
dataset.isnull().sum()
```

```python
x=dataset.drop(columns='SalePrice',axis=1)
y=dataset['SalePrice']
```

```python
print(x)
print(x.shape)
```

```python
print(y)
print(y.shape)
```

```python
x = pd.get_dummies(x)
```

```python
#splitting the data into train and test
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
model=LinearRegression()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
```

```python
from sklearn.impute import SimpleImputer
pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),  # Handles missing values
    ('regressor', LinearRegression())             # Trains the model
])

pipeline.fit(x_train, y_train)
```

```python
#accuracy score
#evaluating
x_train_prediction=pipeline.predict(x_train)
mse = mean_squared_error(y_train, x_train_prediction)
rmse = np.sqrt(mse)

r2 = r2_score(y_train, x_train_prediction)

print(f"Training RMSE: {rmse:.2f}")
print(f"Training R² Score: {r2:.4f}")
```

```python
x_test_prediction = pipeline.predict(x_test)
mse = mean_squared_error(y_test, x_test_prediction)
rmse = np.sqrt(mse)

r2 = r2_score(y_test, x_test_prediction)
print(f"Test RMSE: {rmse:.2f}")
print(f"Test R² Score: {r2:.4f}")
```

```python
regressor = pipeline.named_steps['regressor']
print("Intercept:", regressor.intercept_)
print("Coefficient:", regressor.coef_[0])


```

```python
plt.figure(figsize=(8, 5))

#actual values
plt.scatter(range(len(y_test)), y_test, color='red', label='Actual')

#predicted values
plt.scatter(range(len(x_test_prediction)), x_test_prediction, color='blue', label='Predicted', marker='x')

plt.title("Actual vs Predicted Scores")
plt.xlabel("Test Sample Index")
plt.ylabel("Score")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(pipeline, x_train, y_train, cv=5, scoring='r2')
print("Cross-validation R² scores:", scores)
print("Mean CV R²:", scores.mean())
```

```python
print(" Model completed successfully.")
print(" Training R² Score:", round(r2_score(y_train, x_train_prediction), 4))
print(" Test R² Score:", round(r2_score(y_test, x_test_prediction), 4))
print(" Cross-Validation Mean R²:", round(scores.mean(), 4))
print(" The model performs well and is ready for deployment or further tuning.")
```
